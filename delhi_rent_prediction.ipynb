```python
data['localityName'] = data['localityName'].apply(lambda x: x.strip())
location_count = data['localityName'].value_counts()
```

```python
location_count_less_than_10 = location_count[location_count <=10]
location_count_less_than_10
```

```python
data['localityName'] = data['localityName'].apply(lambda x: 'other' if x in location_count_less_than_10 else x)
```

```python
data[data['price']<1300]
```

```python
upper_limit = data['bedrooms'].quantile(0.9992)
upper_limit
```

```python
lower_limit = data['bedrooms'].quantile(0.0008)
lower_limit
```

```python
new_data = data[(data['bedrooms'] < 8) & (data['bedrooms'] > 1)]
```

```python
new_data.describe()
```

```python
new_data['price_per_sqft'] = new_data['price']/new_data['size_sq_ft']
```

```python
new_data
```

```python
new_data.to_csv('cleaned_data')
```

```python
X = new_data.drop(columns=['price','propertyType'])
y = new_data['price']
```

```python
X.head()
```

```python
y.head()
```

```python
#model training with linear regression
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse=False,handle_unknown='ignore'),[2]), remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
yp_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,yp_lr)
```
